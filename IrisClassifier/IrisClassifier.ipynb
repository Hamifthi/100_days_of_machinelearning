```python
import numpy as np
import sklearn
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
normalizer = sklearn.preprocessing.StandardScaler()
x = normalizer.fit_transform(x)
```

```python
from sklearn import linear_model
def linearModel(x, y):
    model = linear_model.LinearRegression()
    model.fit(x, y)
    return model.score(x, y)
```

```python
import tensorflow as tf
from tensorflow import keras
def singleNeuralNetModel(x, y):
    y = keras.utils.to_categorical(y)
    model = tf.keras.Sequential()
    model.add(keras.layers.Dense(8, activation = 'relu', input_dim = 4))
    model.add(keras.layers.Dense(3, activation = 'softmax'))
    model.compile(loss = 'mean_squared_error', optimizer = 'sgd', metrics=['accuracy'])
    model.fit(x, y, epochs = 100, batch_size = 5)
    return model.evaluate(x, y)
```

```python
print('performance for linear model is {}'.format(linearModel(x, y)))
print('performance for single neural net model is {}'.format(singleNeuralNetModel(x, y)))
```
